### Set Up "Gold Standard" Test Set

```python
import pandas as pd
import openpyxl
```

```python
#Pull in parquet file 
df_base = pd.read_parquet('data/LRC_articles.parquet')
df_base.head()
```

```python
# Select articles for gold standard/manual labeling
df_base['year_month'] = pd.to_datetime(df_base['date']).dt.to_period('M')

# Sample n articles per outlet-month combination
n=1 
gold_standard_uuids = (df_base.groupby(['outlet_name', 'year_month'])
                         .sample(n, random_state=3)['uuid'].tolist())

#print(len(gold_standard_uuids)) #n=1, 756

gold_standard_df = df_base[df_base['uuid'].isin(gold_standard_uuids)]
gold_standard_df = gold_standard_df[['uuid', 'url', 'outlet_name', 'bias', 'date', 'content']]
```

```python
# output df to review labels
gold_standard_df.to_excel('data/gold_standard_for_labeling2.xlsx', index=False)
```

```python
# Read back in what has been labeled, add labels to df
labeled_gold_standard_df = pd.read_excel('data/gold_standard_labeled.xlsx').fillna(0)
labeled_gold_standard_df.head()

#list of uuids that were labeled
labeled_uuids = labeled_gold_standard_df[labeled_gold_standard_df['# Frames Labeled'] > 0]['uuid'].tolist()
```
